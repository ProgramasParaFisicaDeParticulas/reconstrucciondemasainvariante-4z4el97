```python
#panda nos permite manipular el archivo csv https://pandas.pydata.org
import pandas as pd
#numpy nos permite hacer operaciones usando vectores https://numpy.org
import numpy as np
#plotly nos permite crear graficas interactivas https://plotly.com/python/
import plotly.express as px
%matplotlib inline
#usamos la liga que está directamente en la página.
particles = pd.read_csv('https://opendata.cern.ch/record/5208/files/Zmumu.csv',delimiter=',')
particles.head()
```

```python
fig_5=px.histogram(particles['pt1'],particles['pt1'])
fig_5.show()
```

```python
# filtrar eventos donde ambas cargas no sean positivas ni negativas
z_cand = particles[~((particles['Q1'] == 1)&(particles['Q2'] == 1))& ~((particles['Q1'] == -1)&(particles['Q2'] == -1))]
cantidad_datos = z_cand.shape[0]
print("Número de datos después de aplicar el filtro adicional:", cantidad_datos)
z_cand.head()

Carga = particles['Q1'] + particles['Q2']
fig_6=px.histogram(Carga, labels={'value':'Carga' , 'y':'Frecuencia'}, title='Suma de las cargas de ambos muones', color_discrete_sequence= ['blue'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```

```python
#Filtrar eventos con iso
z_cand_iso_cero = z_cand[(z_cand['iso1']==0) &(z_cand['iso2']==0)]
cantidad_datos_iso_cero = z_cand_iso_cero.shape[0]
print("Número de datos después de aplicar el filtro adicional:", cantidad_datos_iso_cero)
z_cand_iso_cero.head()
```

```python
masas_invariantes_iso_cero = np.sqrt(2*z_cand_iso_cero['pt1']*z_cand_iso_cero['pt2']*(np.cosh(z_cand_iso_cero['eta1']-z_cand_iso_cero['eta2'])-np.cos(z_cand_iso_cero['phi1']-z_cand_iso_cero['phi2'])))

fig_6=px.histogram(masas_invariantes_iso_cero, labels={'value':'Masa Invariante (Gev)' , 'y':'Frecuencia'}, title='Histograma de Masas Invariantes iso=0', color_discrete_sequence= ['blue'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```

```python
fig_6=px.histogram(z_cand_iso_cero['pt1'], labels={'value':'Momento del primer muon' , 'y':'Frecuencia'}, title='Momento del primer muon iso=0', color_discrete_sequence= ['blue'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```

```python
fig_6=px.histogram(z_cand_iso_cero['pt2'], labels={'value':'Momento del segundo muon' , 'y':'Frecuencia'}, title='Momento del segundo muon iso=0', color_discrete_sequence= ['green'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```

```python
fig_6=px.histogram(z_cand_iso_cero['eta1'], labels={'value':'eta1 del primer muon' , 'y':'Frecuencia'}, title='Pseudorapidez del primer muon iso=0', color_discrete_sequence= ['blue'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```

```python
fig_6=px.histogram(z_cand_iso_cero['eta2'], labels={'value':'eta2 del segundo muon' , 'y':'Frecuencia'}, title='Pseudorapidez del segundo muon iso=0', color_discrete_sequence= ['green'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```

```python
fig_6=px.histogram(z_cand_iso_cero['phi1'], labels={'value':'phi1 del primer muon' , 'y':'Frecuencia'}, title='Ángulo phi del primer muon iso=0', color_discrete_sequence= ['blue'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```

```python
fig_6=px.histogram(z_cand_iso_cero['phi2'], labels={'value':'phi2 del segundo muon' , 'y':'Frecuencia'}, title='Ángulo phi del segundo muon iso=0', color_discrete_sequence= ['green'])
fig_6.update_yaxes(title_text='Frecuencia')
fig_6.show()
```
